## **Descripción del problema a solucionar**

El programa a continuación genera en una carpeta especificada 1000 archivos de tipo Excel, Word, TXT, CSV, JSON, XML, PowerPoint, HTML y PNG en múltiples carpetas creadas todas de manera aleatoria. El reto si decides aceptarlo es recorrer el árbol de carpetas e identificar los tipos de archivos, crear un nuevo directorio organizado y clasificarlos por tipo (mover los archivos), creando una carpeta llamada Excel para los archivos tipo Excel, una carpeta llamada Word para los archivos tipo Word y así sucesivamente para los demás tipos de archivos. Los archivos en la nueva carpeta deben ser renombrados usando un consecutivo luego de ordenar los archivos por nombre, cada archivo debe comenzar por un numero de tres dígitos, por ejemplo, el primer archivo se nombrará “001-Nombre.extension”. Adicionalmente debes guardar en un Pandas Dataframe el nombre del archivo anterior, nombre del archivo actual, la ruta anterior, la nueva ruta, el tipo de archivo, el tamaño en múltiples unidades de medida por columna usando las siguientes unidades, bit, byte, Kilobyte y Megabyte (crear una columna para cada tipo de unidad de medida). Adicionalmente en una columna detallar la cantidad de vocales que tiene el nombre del documento, en una columna detallar el número de consonantes que tiene el nombre del documento, la fecha de creación, la última fecha de modificación y por último la cantidad de archivos del mismo tipo en la carpeta de su mismo tipo. Este programa debe guardar su procedimiento en un log de eventos (sea tan preciso como pueda) el log debe tener: fecha, hora, minuto, segundo y milisegundo un separador tabular, la instrucción a realizar, y el tiempo que demora en realizar esta operación. El Log debe tener de encabezado, el nombre del usuario que ejecuta el procedimiento, el sistema operativo, la plataforma y todas sus posibles descripciones, al final detallar la cantidad de procedimientos realizados y el tipo de acción, por ejemplo, contar archivos es una acción, mover es una segunda acción, renombrar es una acción etc.

```python
import os
from openpyxl import Workbook
from docx import Document
import random as rnd
import pandas as pd
from PIL import Image, ImageDraw, ImageFont
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import time
import datetime
import warnings
warnings.filterwarnings('ignore')
import logging
import tqdm
from google.colab import files

inicio = time.time()
print('Ingresar la cantidad de archivos a procesar, ingresar solo numeros enteros.')
print('Si ingresa el numero 10, por ejemplo, se crearan 10 archivos de cada uno.')
print('Generando un total de 100 archivos, A continuacion ingresar el numero.')
archivos = int(input('Ingresar la cantidad: '))
hoy = datetime.date.today().strftime('%Y%m%d')
nombre_archivo_log = f"log_{hoy}.log"
logging.basicConfig(filename=nombre_archivo_log, level=logging.INFO,
                    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')
logging.info("Iniciando el proceso, por CastilloEnterprises (^人^)(^人^)(^人^)")


#Creamos el directorio (carpeta) en donde se crearan los archivos
DirectorioActual = os.getcwd()
print(f'El directorio actual de trabajo es: \n\t--> {DirectorioActual}, \nEsta carpeta contendrá los archivos del trabajo final')
CarpetaNueva = "CarpetaArchivosTrabajoFinal"
os.makedirs(CarpetaNueva, exist_ok=True)
logging.info("Se crea el directorio {}".format(CarpetaNueva))
print(f"'{CarpetaNueva}' La carpeeta ha sido creada.")
# Nombre de la carpeta donde se crearán los archivos
carpeta = os.path.join(DirectorioActual, "CarpetaArchivosTrabajoFinal")
logging.info("La ruta de trabajo será {}".format(carpeta))
RutaNombres = r'https://raw.githubusercontent.com/juliancastillo-udea/Curso_AlgoritmosProgramacion/main/TrabajoFinal/NombresArgentina.csv'
logging.info("Cargando CSV con nombres")
dfNombres = pd.read_csv(RutaNombres, encoding='ISO-8859-1')
Nombres = dfNombres['name'].tolist()
logging.info("Reemplazando nombres y detalles del documento")
for i in tqdm.trange(len(Nombres)):
    if ' ' in Nombres[i]:
        Nombres[i]=Nombres[i].replace(' ', '_')
logging.info("Creacion de archivos")
for i in tqdm.trange(int(archivos), miniters=int(1), ascii=True,desc="Creando Archivos", dynamic_ncols=True):
    # Crear un archivo de Excel usando openpyxl--------------------------------
    logging.info(f"Creacion de archivos-->Excel {i+1}")
    wb = Workbook()
    ws = wb.active
    ws['A1'] = 'Hola soy un numero aleatorio'
    ws['A2'] = rnd.random()
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    excel_filename = os.path.join(carpeta, tempNombre+'.xlsx')
    wb.save(excel_filename)

    # Crear un archivo de Word usando python-docx------------------------------
    logging.info(f"Creacion de archivos-->Word {i+1}")
    doc = Document()
    doc.add_paragraph(f"¡Hola! Este es un archivo de Word creado desde Python. \n Soy un numero aleatorio {rnd.random()}")
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    word_filename = os.path.join(carpeta, tempNombre+'.docx')
    doc.save(word_filename)

    # Crear un archivo de texto (txt)------------------------------------------
    logging.info(f"Creacion de archivos-->Texto {i+1}")
    texto = f"¡Hola! Este es un archivo de Word creado desde Python. \n Soy un numero aleatorio {rnd.random()}"
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    txt_filename = os.path.join(carpeta, tempNombre+'.txt')
    with open(txt_filename, 'w') as txt_file:
        txt_file.write(texto)

    # Crear un archivo CSV-----------------------------------------------------
    logging.info(f"Creacion de archivos-->CSV {i+1}")
    csv_data = f"Nombre,Edad\n{rnd.choice(Nombres)},{rnd.uniform(18,60)}\n"
    csv_data += f"{rnd.choice(Nombres)},{rnd.uniform(18,60)}\n"
    csv_data += f"{rnd.choice(Nombres)},{rnd.uniform(18,60)}\n"
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    csv_filename = os.path.join(carpeta, tempNombre+'.csv')
    with open(csv_filename, 'w') as csv_file:
        csv_file.write(csv_data)

    # Crear un archivo JSON----------------------------------------------------
    logging.info(f"Creacion de archivos-->JSON {i+1}")
    import json
    json_data = {"nombre": rnd.choice(Nombres), "edad": rnd.uniform(18,60)}
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    json_filename = os.path.join(carpeta, tempNombre+'.json')
    with open(json_filename, 'w') as json_file:
        json.dump(json_data, json_file, indent=4)

    # Crear un archivo XML-----------------------------------------------------
    logging.info(f"Creacion de archivos-->XML {i+1}")
    xml_data = f"""<?xml version="1.0" encoding="UTF-8"?>
    <datos>
        <persona>
            <nombre>{rnd.choice(Nombres)}</nombre>
            <edad>{rnd.uniform(18,60)}</edad>
        </persona>
        <persona>
            <nombre>{rnd.choice(Nombres)}</nombre>
            <edad>{rnd.uniform(18,60)}</edad>
        </persona>
    </datos>
    """
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    xml_filename = os.path.join(carpeta, tempNombre+'.xml')
    with open(xml_filename, 'w') as xml_file:
        xml_file.write(xml_data)

    # Crear un archivo de PowerPoint-------------------------------------------
    logging.info(f"Creacion de archivos-->PowerPoint {i+1}")
    from pptx import Presentation
    ppt = Presentation()
    slide = ppt.slides.add_slide(ppt.slide_layouts[0])
    slide.shapes.title.text = f"Presentación creada desde Python,\n Soy un numero aleatorio {rnd.random()}"
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    ppt_filename = os.path.join(carpeta, tempNombre+'.pptx')
    ppt.save(ppt_filename)

    # Crear un archivo HTML----------------------------------------------------
    logging.info(f"Creacion de archivos-->HTML {i+1}")
    html_data = """<!DOCTYPE html>
    <html>
    <head>
        <title>Página HTML creada desde Python</title>
    </head>
    <body>
        <h1>Hola desde Python</h1>
        <h1>Hola soy un numero aleatorio {}</h1>
    </body>
    </html>
    """.format(rnd.random())
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    html_filename = os.path.join(carpeta, tempNombre+'.html')
    with open(html_filename, 'w') as html_file:
        html_file.write(html_data)

    #Crear una imagen vacía----------------------------------------------------
    logging.info(f"Creacion de archivos-->Imagen PNG {i+1}")
    ancho = 400
    alto = 400
    imagen_vacia = Image.new('RGB', (ancho, alto), 'white')
    texto = f"Hola soy un numero aleatorio\n {rnd.random()}"
    #fuente = ImageFont.truetype("arial.ttf", 20)
    dibujar = ImageDraw.Draw(imagen_vacia)
    bbox = dibujar.textbbox((0,0), texto)
    ancho_texto = bbox[2] - bbox[0]
    alto_texto = bbox[3] - bbox[1]
    x = (ancho - ancho_texto) / 2
    y = (alto - alto_texto) / 2
    dibujar.text((x, y), texto, fill="black")
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    imagen = os.path.join(carpeta, tempNombre+'.png')
    imagen_vacia.save(imagen)

    #Archivo PDF---------------------------------------------------------------
    logging.info(f"Creacion de archivos-->PDF {i+1}")
    tempNombre = rnd.choice(Nombres)
    Nombres.remove(tempNombre)
    pdfarchivo = os.path.join(carpeta, tempNombre+'.pdf')
    c = canvas.Canvas(pdfarchivo, pagesize=letter)
    width, height = letter
    texto = f"Hola soy un archivo PDF con un numero aleatorio \n {rnd.random()}"
    c.drawString(width/4 - 15, height/3, texto)  # Ajusta la posición según lo que necesites
    c.save()
logging.info("Proceso Terminado......")
logging.info("Gracias por usar Castillo Enterprises")
logging.shutdown()

fin = time.time()
tiempoejecucion = fin - inicio
minutos = int(tiempoejecucion // 60)
segundos = tiempoejecucion % 60
print(f"El script tomó {minutos} minutos y {segundos:.2f} segundos para ejecutarse.")
```

```python
from google.colab import drive
import os
import shutil
import pandas as pd
from datetime import datetime
import time
import zipfile

drive.mount('/content/drive')

class OrganizadorDocumentos:
    def __init__(self, ruta_principal, ruta_destino, log_path):
        self.ruta_principal = ruta_principal
        self.ruta_destino = ruta_destino
        self.log_path = log_path
        self.tipos_archivos_validos = {'xlsx', 'xls', 'docx', 'doc', 'txt', 'csv', 'json', 'xml', 'pptx', 'html', 'png'}
        self.registro_eventos = []

    def organizar_documentos(self):
        if not os.path.exists(self.ruta_destino):
            os.makedirs(self.ruta_destino)

        for carpeta_actual, _, archivos in os.walk(self.ruta_principal):
            for archivo in archivos:
                tipo_archivo = archivo.split('.')[-1].lower()

                if tipo_archivo in self.tipos_archivos_validos:
                    ruta_original = os.path.join(carpeta_actual, archivo)

                    carpeta_destino = os.path.join(self.ruta_destino, tipo_archivo.capitalize())
                    if not os.path.exists(carpeta_destino):
                        os.makedirs(carpeta_destino)

                    nuevo_nombre = f"{str(len(os.listdir(carpeta_destino)) + 1).zfill(3)}-{archivo}"
                    nueva_ruta = os.path.join(carpeta_destino, nuevo_nombre)
                    shutil.move(ruta_original, nueva_ruta)

                    self.registrar_evento(archivo, nuevo_nombre, ruta_original, nueva_ruta, tipo_archivo)

        self.actualizar_log()

    def explorar_carpetas(self):
        for carpeta_actual, carpetas, archivos in os.walk(self.ruta_principal):
            print(f"\nCarpeta: {carpeta_actual}")

            # Mostrar carpetas en la carpeta actual
            if carpetas:
                print("Carpetas:")
                for carpeta in carpetas:
                    print(f"  - {carpeta}")

            # Mostrar archivos en la carpeta actual
            if archivos:
                print("Archivos:")
                for archivo in archivos:
                    print(f"  - {archivo}")

    def buscar_documento(self, nombre_documento):
        encontrado = False
        for carpeta_actual, _, archivos in os.walk(self.ruta_destino):
            for archivo in archivos:
                if nombre_documento.lower() in archivo.lower():
                    print(f"Documento encontrado: {os.path.join(carpeta_actual, archivo)}")
                    encontrado = True
        return encontrado

    def buscar_en_carpeta(self, carpeta, nombre_documento):
        encontrado = False
        for carpeta_actual, _, archivos in os.walk(carpeta):
            for archivo in archivos:
                if nombre_documento.lower() in archivo.lower():
                    print(f"Documento encontrado en {carpeta_actual}: {os.path.join(carpeta_actual, archivo)}")
                    encontrado = True
        return encontrado

    def buscar_en_tipo_archivo(self, tipo_archivo, nombre_documento):
        carpeta_tipo_archivo = os.path.join(self.ruta_destino, tipo_archivo.capitalize())
        return self.buscar_en_carpeta(carpeta_tipo_archivo, nombre_documento)

    def listar_nombres_en_tipo_archivo(self, tipo_archivo):
        carpeta_tipo_archivo = os.path.join(self.ruta_destino, tipo_archivo.capitalize())
        nombres_archivos = [archivo for archivo in os.listdir(carpeta_tipo_archivo) if os.path.isfile(os.path.join(carpeta_tipo_archivo, archivo))]
        # Agregar nombres de archivos en archivos .zip
        for archivo_zip in [archivo for archivo in os.listdir(carpeta_tipo_archivo) if archivo.lower().endswith('.zip')]:
            with zipfile.ZipFile(os.path.join(carpeta_tipo_archivo, archivo_zip), 'r') as zip_ref:
                nombres_archivos.extend(zip_ref.namelist())
        return nombres_archivos

    def mostrar_registro_eventos(self):
        # Cargar datos del registro en un DataFrame de Pandas
        columnas = ["Fecha", "Nombre Anterior", "Nombre Actual", "Ruta Anterior", "Ruta Nueva", "Tipo Archivo",
                    "Tamaño (bytes)", "Tamaño (KB)", "Tamaño (MB)", "Vocales", "Consonantes", "Fecha Creación",
                    "Fecha Modificación", "Cantidad Archivos Tipo"]

        df_registro = pd.read_csv(self.log_path, sep='\t', header=0, names=columnas)

        # Mostrar el DataFrame
        print("Registro de Eventos:")
        print(df_registro)

    def registrar_evento(self, nombre_anterior, nombre_actual, ruta_anterior, nueva_ruta, tipo_archivo):
        tiempo_actual = time.time()
        fecha_hora = datetime.fromtimestamp(tiempo_actual).strftime('%Y-%m-%d %H:%M:%S.%f')[:-3]

        tamano = os.path.getsize(nueva_ruta)
        tamano_kb = tamano / 1024
        tamano_mb = tamano_kb / 1024

        vocales = sum(1 for char in nombre_actual if char.lower() in 'aeiou')
        consonantes = sum(1 for char in nombre_actual if char.isalpha() and char.lower() not in 'aeiou')

        fecha_creacion = datetime.fromtimestamp(os.path.getctime(nueva_ruta)).strftime('%Y-%m-%d %H:%M:%S')
        fecha_modificacion = datetime.fromtimestamp(os.path.getmtime(nueva_ruta)).strftime('%Y-%m-%d %H:%M:%S')

        cantidad_archivos_tipo = len([f for f in os.listdir(os.path.dirname(nueva_ruta)) if f.lower().endswith(f".{tipo_archivo}")])

        self.registro_eventos.append([fecha_hora, nombre_anterior, nombre_actual, ruta_anterior, nueva_ruta, tipo_archivo,
                                     tamano, tamano_kb, tamano_mb, vocales, consonantes, fecha_creacion, fecha_modificacion,
                                     cantidad_archivos_tipo])

    def actualizar_log(self):
        with open(self.log_path, 'a') as log:
            log.write(f"Fecha\t\t\tNombre Anterior\tNombre Actual\tRuta Anterior\t\tRuta Nueva\t\tTipo Archivo\t"
                      f"Tamaño (bytes)\tTamaño (KB)\tTamaño (MB)\tVocales\tConsonantes\tFecha Creación\tFecha Modificación\t"
                      f"Cantidad Archivos Tipo\n")
            for evento in self.registro_eventos:
                log.write('\t'.join(map(str, evento)) + '\n')

# Definir rutas en Google Drive
ruta_principal_drive = "/content/drive/MyDrive/Colab Notebooks/TrabajoFinalAlgoritmia/Origen/CarpetaArchivosTrabajoFinal (1)/content/CarpetaArchivosTrabajoFinal"
ruta_destino_drive = "/content/drive/MyDrive/Colab Notebooks/TrabajoFinalAlgoritmia/Destino"
log_path_drive = "/content/drive/MyDrive/Colab Notebooks/TrabajoFinalAlgoritmia/Log/log.txt"

# Ejemplo de uso
organizador = OrganizadorDocumentos(ruta_principal_drive, ruta_destino_drive, log_path_drive)
organizador.organizar_documentos()
organizador.explorar_carpetas()
organizador.buscar_documento("001")

# Bucle para realizar búsquedas adicionales
while True:
    buscar_otra_vez = input("¿Quieres buscar otro documento? (Escribe (si) o (no): ").lower()
    if buscar_otra_vez != 'si':
        break

    # Mostrar los tipos de archivos disponibles
    print("Tipos de archivos disponibles:")
    for tipo_archivo in organizador.tipos_archivos_validos:
        print(f"  - {tipo_archivo}")

    tipo_archivo = input("Ingresa el tipo de archivo (por ejemplo, 'csv'): ").lower()
    nombre_documento = input("Ingresa el nombre del documento que estás buscando: ")

    # Buscar en la carpeta general y luego en la carpeta específica del tipo de archivo
    encontrado_general = organizador.buscar_documento(nombre_documento)
    encontrado_tipo_archivo = organizador.buscar_en_tipo_archivo(tipo_archivo, nombre_documento)

    if not (encontrado_general or encontrado_tipo_archivo):
        print("Error, nombre no encontrado.")
        lista_nombres_tipo_archivo = organizador.listar_nombres_en_tipo_archivo(tipo_archivo)
        print(f"Nombres disponibles en la carpeta {tipo_archivo.capitalize()}:")
        print("\n".join(lista_nombres_tipo_archivo))
    else:
        # Mostrar el registro de eventos después de la búsqueda
        organizador.mostrar_registro_eventos()
```
